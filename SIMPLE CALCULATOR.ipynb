```python
from tkinter import*

root=Tk()
root.title("Simple Calculator")
e=Entry(root,width=45,borderwidth=4)
e.grid(row=0,column=0,columnspan=3)

def ButtonClick(Number):
    current=e.get()
    e.delete(0,END)
    e.insert(0,str(current)+str(Number))

def Buttonclear():
    e.delete(0,END)

def Buttonadd():
    global fnum 
    global math
    math=1
    first_num=e.get()
    e.delete(0,END)
    fnum=int(first_num)

def Buttonsub():
    global fnum
    global math
    math=2
    first_num=e.get()
    e.delete(0,END)
    fnum=int(first_num)

def Buttonmulti():
    global fnum
    global math
    math=3 
    first_num=e.get()
    e.delete(0,END)
    fnum=int(first_num)

def Buttondiv():
    global fnum
    global math
    math=4 
    first_num=e.get()
    e.delete(0,END)
    fnum=int(first_num)


def Buttonequal():
    second_num=e.get()
    e.delete(0,END)

    if math==1:
        e.insert(0,fnum+int(second_num))
    elif math==2:
        e.insert(0,fnum-int(second_num))
    elif math==3:
        e.insert(0,fnum*int(second_num))
    else:
        e.insert(0,fnum//int(second_num))



button1=Button(root,text=1,padx=45,pady=20,command=lambda : ButtonClick(1)).grid(row=1,column=0)
button2=Button(root,text=2,padx=45,pady=20,command=lambda : ButtonClick(2)).grid(row=1,column=1)
button3=Button(root,text=3,padx=45,pady=20,command=lambda : ButtonClick(3)).grid(row=1,column=2)

button4=Button(root,text=4,padx=45,pady=20,command=lambda : ButtonClick(4)).grid(row=2,column=0)
button5=Button(root,text=5,padx=45,pady=20,command=lambda : ButtonClick(5)).grid(row=2,column=1)
button6=Button(root,text=6,padx=45,pady=20,command=lambda : ButtonClick(6)).grid(row=2,column=2)

button7=Button(root,text=7,padx=45,pady=20,command=lambda : ButtonClick(7)).grid(row=3,column=0)
button8=Button(root,text=8,padx=45,pady=20,command=lambda : ButtonClick(8)).grid(row=3,column=1)
button9=Button(root,text=9,padx=45,pady=20,command=lambda : ButtonClick(9)).grid(row=3,column=2)
button0=Button(root,text=0,padx=45,pady=20,command=lambda : ButtonClick(0)).grid(row=4,column=0)

buttonequal=Button(root,text="=",padx=100,pady=20,command=Buttonequal).grid(row=4,column=1,columnspan=2)
buttonadd=Button(root,text="+",padx=45,pady=20,command=Buttonadd).grid(row=5,column=0)
buttonsub=Button(root,text="-",padx=45,pady=20,command=Buttonsub).grid(row=5,column=1)
buttonmultiply=Button(root,text="*",padx=45,pady=20,command=Buttonmulti).grid(row=5,column=2)
buttondiv=Button(root,text="/",padx=45,pady=20,command=Buttondiv).grid(row=6,column=0)
buttonclear=Button(root,text="Clear",padx=95,pady=20,command=Buttonclear).grid(row=6,column=1,columnspan=2)





root.mainloop()
```
